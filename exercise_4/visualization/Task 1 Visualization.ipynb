### Connected components

```python
import numpy as np
import matplotlib.pyplot as plt

ten_clusters_setting = [
    [1, 7, 4, 4, 2, 2, 2, 4, 2, 3, 1],
    [1, 2, 3, 2, 3, 3, 5, 2, 1, 1],
    [1, 2, 3, 2, 2, 2, 4, 3, 2, 1, 2],
    [1, 3, 1, 2, 4, 2, 4, 2, 2],
    [1, 3, 6, 2, 2, 1, 2, 1, 2, 4],
    [1, 2, 4, 4, 4, 1, 3, 2, 3, 3],
    [1, 1, 4, 4, 4, 1, 2, 2, 1, 4, 2],
    [1, 2, 4, 4, 1, 4, 2, 4, 2, 3, 2],
    [1, 3, 4, 2, 2, 1, 2, 4, 1, 2, 1],
    [1, 2, 4, 2, 1, 4, 5, 2, 2, 2, 1]
]

five_clusters_setting = [ 
    [1, 1, 2, 3, 3, 2],
    [1, 1, 2, 2, 4, 2],
    [1, 3, 2, 3, 2, 3],
    [1, 3, 2, 3, 4, 2],
    [1, 6, 4, 5, 4, 3],
    [1, 5, 3, 3, 5, 2],
    [1, 5, 4, 2, 2, 4],
    [1, 3, 3, 2, 2, 2],
    [1, 3, 2, 2, 1, 3],
    [1, 5, 4, 4, 2, 2]
]
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 5, figsize=(30, 8))
for i, ax in enumerate(fig.axes):
    arr = ten_clusters_setting[i]
    print(arr)
    x = np.arange(len(arr))
    ax.title.set_text('Graph {}'.format(str(i)))
    ax.bar(x, arr, width = 0.6)
    ax.set_ylim([0, 8])

# plt.ylim([0, 10])
plt.tight_layout()
plt.savefig('connected_components_10_max.png')
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 5, figsize=(30, 8))
plt.title('Max 5 clusters setting')

for i, ax in enumerate(fig.axes):
    arr = five_clusters_setting[i]
    x = np.arange(len(arr))
    ax.title.set_text('Graph {}'.format(str(i)))
    ax.bar(x, arr, width = 0.6)
    ax.set_ylim([0, 7])

plt.tight_layout()
plt.savefig('connected_components_5_max.png')
```

### Modularity

```python
mod_ten_clusters_setting = [
    -0.017755,
    0.052959,
    0.066122,
    -0.012755,
    0.011531,
    0.073163,
    0.008367,
    -0.022755,
    0.030816,
    -0.044592
]

mod_five_clusters_setting = [
    0.058571,
    0.050816,
    0.024388,
    0.039898,
    -0.056531,
    -0.006837,
    -0.032245,
    0.031531,
    0.016429,
    -0.024286
]
```

```python
fig, ax = plt.subplots(figsize=(20, 5))
x = np.arange(10) # We have 10 runs
plt.plot(x, mod_ten_clusters_setting, marker='o', markerfacecolor='tomato', markersize=12, color='salmon', linewidth=4)
plt.plot( x, mod_five_clusters_setting, marker='x', markerfacecolor='dodgerblue', markersize=12, color='deepskyblue', linewidth=2)
plt.legend(['Max 10 clusters setting', 'Max 5 clusters setting'], prop={'size': 14})
ax.set_title('Modularity of graphs for different settings')
ax.set_ylabel('modularity')
ax.set_xlabel('run number')
ax.set_xticks(x)
plt.axhline(0, color='black')
plt.tight_layout()
plt.savefig('modularity.png')
plt.show()
```
