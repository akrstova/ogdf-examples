### Width and Height of Graph Drawings

```python
import os
import xml.etree.ElementTree as ET
```

```python
plt_width = 0.4
dirs = ['longestpath_splitheuristic', 'optimalranking_medianheuristic', 'longestpath_medianheuristic', 'optimalranking_splitheuristic']
for params_combo in dirs:  
    files = list(map(lambda x: 'graph_output/' + params_combo + '/' + x + '.svg', list(running_times["optimal_split"].keys())))
    names = list(running_times['optimal_split'].keys())
    widths = []
    heights = []
    for filename in files:
        tree = ET.parse(filename).getroot()
        viewBox = tree.attrib["viewBox"]
        x0, y0, x1, y1 = viewBox.split(' ')
        width = float(x1) - float(x0)
        height = float(y1) - float(y0)
#         print('Image {} W: {}, H: {}'.format(filename, width, height))
        widths.append(width)
        heights.append(height)
        
    fig, ax = plt.subplots(figsize=(20,5))
    pos = list(range(len(files)))
    plt.bar(pos, widths, plt_width, alpha=0.5, color='#e38819', label='widths')
    plt.bar([p + plt_width for p in pos], heights, plt_width, alpha=.5, color='#19cbe3', label='heights')
    ax.set_title('Image width vs image height for combination: {}'.format(params_combo))
    ax.set_ylabel('pixels')
    ax.set_xticks([p + 1.5 * plt_width for p in pos])
    ax.set_xticklabels( names, rotation='vertical')
    plt.legend(['width', 'height'], loc='upper left')

    plt.tight_layout()
    plt.savefig('{}_w_h.png'.format(params_combo))
    plt.show()
```
